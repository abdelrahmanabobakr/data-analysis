```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\\New folder\\mony\\data analysis\\amazon.csv")
```

```python
df.info()
```

```python
df.product_id.value_counts()
```

```python
df.duplicated().sum()
```

```python
df['rating'] = pd.to_numeric(df['rating'].str.replace('|', ''))
```

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'].str.replace(r'[^0-9.]', '', regex=True))
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(r'[₹$,%]', '', regex=True))
```

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace(r'[₹$,%]', '', regex=True))
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace(r'[₹$,%]', '', regex=True))
```

```python
df.info()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(data=df, x='actual_price', y='rating')
plt.title('Relationship between Actual Price and Ratings')
plt.show()
```

```python
sns.scatterplot(data=df, x='discount_percentage', y='rating')
plt.title('Relationship between Actual Price and Ratings')
plt.show()
```

```python
top_rated = df.sort_values(by='rating', ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.barh(top_rated['product_name'], top_rated['rating'], color='skyblue')
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Product Name', fontsize=12)
plt.title('Top 10 Highest Rated Products', fontsize=14)
plt.gca().invert_yaxis()  
plt.grid(axis='x')
plt.tight_layout()  
plt.show()
```

```python
most_discounted = df.sort_values(by='discount_percentage', ascending=False).head(10)
print(most_discounted[['product_name', 'discount_percentage']])
```

```python
max_price = df['actual_price'].max()
high_price_product = df[df['actual_price'] == max_price]

high_price_product_info = high_price_product[['product_name', 'actual_price', 'rating', 'rating_count']]

print(high_price_product_info)
```

```python

user_purchases = df.groupby('user_id')['actual_price'].sum().reset_index()
user_purchases.rename(columns={'actual_price': 'total_purchase_amount'}, inplace=True)

top_users = user_purchases.sort_values(by='total_purchase_amount', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_users['user_id'], top_users['total_purchase_amount'], color='skyblue')
plt.xlabel('Total Purchase Amount')
plt.ylabel('User ID', fontsize=12)
plt.title('Top 10 Users by Total Purchases')
plt.gca().invert_yaxis()  
plt.grid(axis='x')
plt.show()
```

```python
sns.histplot(df['rating'], kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
